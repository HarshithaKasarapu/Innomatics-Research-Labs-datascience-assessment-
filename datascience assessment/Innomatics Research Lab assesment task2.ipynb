```python
##finding outliers
def outliers_iqr(vardata,col):    
    q1 = vardata[col].quantile(0.25)
    q3 = vardata[col].quantile(0.75)
    iqr = q3 - q1

    lb = q1 - (1.5*iqr)
    ub = q3 + (1.5*iqr)
    outliers = []
    for val in vardata[col]:
        if val < lb or val > ub:
            outliers.append(val)

    return outliers,len(outliers)
```

```python
outliers_iqr(vardata, 'X')
```
